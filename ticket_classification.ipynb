```python
# Import libraries
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# Create dataset directly
data = {
    "Ticket": [
        "My payment failed during checkout",
        "Unable to login to my account",
        "App crashes when opening",
        "Need refund for my order",
        "Forgot my password",
        "Website loading very slowly",
        "Payment deducted twice",
        "Cannot reset password",
        "Error while uploading file",
        "Want to update billing info"
    ],

    "Category": [
        "Billing",
        "Account",
        "Technical",
        "Billing",
        "Account",
        "Technical",
        "Billing",
        "Account",
        "Technical",
        "Billing"
    ]
}

# Convert into dataframe
df = pd.DataFrame(data)

# Input and output
X = df["Ticket"]
y = df["Category"]

# Convert text to numbers
vectorizer = CountVectorizer()

X_vectorized = vectorizer.fit_transform(X)

# Train model
model = MultinomialNB()

model.fit(X_vectorized, y)

# Test prediction
sample_ticket = ["Payment issue in my account"]

sample_vector = vectorizer.transform(sample_ticket)

prediction = model.predict(sample_vector)

# Output
print("Sample Ticket:", sample_ticket[0])

print("Predicted Category:", prediction[0])
```
